```python
import pandas as pd
tip=pd.read_csv('yelp_tip.csv')
tip.head()
businesses = pd.read_csv('yelp_business.csv')
restaurants = businesses[businesses['categories'].str.contains("Restaurants")]
restaurant_tip=tip.merge(restaurants, left_on='business_id', right_on='business_id', how = 'inner')#filtering only restaurants
tip.columns
restaurant_tip.columns
restaurant_tip
```

```python
#len(restraunt_tip)
restaurant_tip=restaurant_tip[restaurant_tip.is_open!=0]#deleting entries for which restaurants are closed
#len(restaurant_tip)
restaurant_tip.columns
restaurant_tip = restaurant_tip.drop(restaurant_tip.columns[[6, 7,11, 12]], axis=1) #dropping unwanted columns
restaurant_tip.head()
```

```python
restaurant_tip['tip_year'] = restaurant_tip['date'].apply(lambda x:int(x[:4]))#getting only year from date
restaurant_tip.head()
```

```python
restaurant_tip[(restaurant_tip['tip_year'] > 2010)]#deleting tips older than 2010
```

```python
restaurant_tip
```

```python
restaurant_tip.text=restaurant_tip.text.astype(str).str.lower()#converting string to lowercase
restaurant_tip
#ind=[i for i in range(len(restaurant_tip))]

#restaurant_tip.set_index('ind')
```

```python
foods=["pasta","pizza","pancake","sandwiches","ice cream","steak","chicken", "biryani"]
for food in foods:
    restaurant_tip[food] = restaurant_tip['text'].str.contains(food).astype('int')
```

```python
grouped = restaurant_tip.groupby(['name'])[foods].apply(lambda x : x.astype(int).sum()).reset_index()
melted = pd.melt(grouped, id_vars=["name"], 
             var_name="food")
melted['food_rank'] = melted.groupby('name')['value'].rank(ascending=False)
melted[melted['food_rank'] <= 3]
```

```python
melted = pd.melt(temp, id_vars=["name"], 
             var_name="cuisines")
melted['cuisine_rank'] = melted.groupby('name')['value'].rank(ascending=False)
```
