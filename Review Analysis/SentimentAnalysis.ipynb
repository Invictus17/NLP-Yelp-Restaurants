## Text Preprocessing 

**Removing stop words**

```python
import nltk
from nltk.corpus import stopwords
# nltk.download('stopwords')

english_stop_words = stopwords.words('english')
def remove_stop_words(corpus):
    removed_stop_words = []
    for review in corpus:
        removed_stop_words.append(
            ' '.join([word for word in review.split() 
                      if word not in english_stop_words])
        )
    return removed_stop_words

relevant_reviews['cleaned_text'] = relevant_reviews['text'].apply(lambda x: remove_stop_words(x))
```

**Stemming/Lemmatization**

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

def get_lemmatized_text(corpus):
    lemmatizer = WordNetLemmatizer()
    return [' '.join([lemmatizer.lemmatize(word) for word in review.split()]) for review in corpus]

relevant_reviews['cleaned_text'] = relevant_reviews['cleaned_text'].apply(lambda x: get_lemmatized_text(x))
```

## Create Train/Validation/Test Split

```python
from sklearn.model_selection import train_test_split
```

```python

X_train_set, X_nottrain, y_train, y_nottrain = train_test_split(relevant_reviews['text'], 
                                                    relevant_reviews['sentiment'], 
                                                    test_size=0.30, 
                                                    random_state=42)

X_test_set, X_val_set, y_test, y_val = train_test_split(X_nottrain, 
                                                    y_nottrain, 
                                                    test_size=0.5, 
                                                    random_state=42)
```

## Feature Engineering

**1-gram Binary**

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(binary=True)
cv.fit(X_train_set)
X_train = cv.transform(X_train_set)
X_test = cv.transform(X_test_set)
X_val = cv.transform(X_val_set)
```

**1-gram Count**

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(binary=False)
cv.fit(X_train)
X_train = cv.transform(X_train_set)
X_test = cv.transform(X_test_set)
X_val = cv.transform(X_val_set)
```

**2-gram Binary**

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(binary=True, ngram_range=(1,2))
cv.fit(X_train)
X_train = cv.transform(X_train_set)
X_test = cv.transform(X_test_set)
X_val = cv.transform(X_val_set)
```

**2-gram Count**

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(binary=False, ngram_range=(1,2))
cv.fit(X_train)
X_train = cv.transform(X_train_set)
X_test = cv.transform(X_test_set)
X_val = cv.transform(X_val_set)
```

**1-gram Tf-Idf**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv = TfidfVectorizer()
cv.fit(X_train)
X_train = cv.transform(X_train_set)
X_test = cv.transform(X_test_set)
X_val = cv.transform(X_val_set)
```

**2-gram Tf-Idf**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv = TfidfVectorizer(ngram_range=(1,2))
cv.fit(X_train)
X_train = cv.transform(X_train_set)
X_test = cv.transform(X_test_set)
X_val = cv.transform(X_val_set)

```

```python
X_test.shape
```

## Multiple Iterations

**Baseline Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

for c in [0.01, 0.05, 0.25, 0.5, 1]:
    
    lr = LogisticRegression(solver = 'lbfgs', 
                            penalty = 'l2', 
                            C=c,
                           max_iter= 300)
    lr.fit(X_train, y_train)
    print ("For C = %s, Accuracy: %s, F1-score: %s" 
           % (c, accuracy_score(y_val, lr.predict(X_val)), f1_score(y_val, lr.predict(X_val))))
```

```python
final_model = LogisticRegression(solver = 'lbfgs', 
                            penalty = 'l2', 
                            C=1,
                           max_iter= 300)
final_model.fit(X_train, y_train)
print ("Final Accuracy: %s, F1-score: %s" 
           % (accuracy_score(y_test, final_model.predict(X_test)), f1_score(y_test, final_model.predict(X_test))))
```

```python
feature_to_coef = {
    word: coef for word, coef in zip(
        cv.get_feature_names(), final_model.coef_[0]
    )
}
print("Best Positive")
for best_positive in sorted(
    feature_to_coef.items(), 
    key=lambda x: x[1], 
    reverse=True)[:15]:
    print (best_positive)

print("Best Negative")

for best_negative in sorted(
    feature_to_coef.items(), 
    key=lambda x: x[1])[:15]:
    print (best_negative)
    
```

**Baseline SVM**

```python
from sklearn.svm import LinearSVC

for c in [0.01, 0.05, 0.25, 0.5, 1]:
    
    svm = LinearSVC(C=c, max_iter=400)
    svm.fit(X_train, y_train)
    print ("For C = %s, Accuracy: %s, F1-score: %s" 
           % (c, accuracy_score(y_val, svm.predict(X_val)), f1_score(y_val, svm.predict(X_val))))


```

```python
final_model = LinearSVC(C=0.25, max_iter=400)

final_model.fit(X_train, y_train)
print ("Final Accuracy: %s, F1-score: %s" 
           % (accuracy_score(y_test, final_model.predict(X_test)), f1_score(y_test, final_model.predict(X_test))))
```
